张量之间的运算——两张量中对应元素进行运算

```python
x=torch.tensor([1.0,2,4,8])
y=torch.tensor([2,2,2,2])
x+y,x-y,x*y,x/y,x**y
```

```python
torch.exp(x)
```

cat将两个张量合在一起，dim=0说明尺寸为0,两张量堆叠起来;dim=1说明尺寸为1,两张量并排堆放

```python
x=torch.arange(12,dtype=torch.float32).reshape(3,4)
y=torch.tensor([[2.0,1,4,3],[1,2,1,2],[4,3,2,1]])
torch.cat((x,y),dim=0),torch.cat((x,y),dim=1)
```

```python
x==y #逻辑运算符构建二维张量
```

```python
x.sum()
```

python中会为新结果分配内存

```python
before=id(y)
y=y+x
id(y)==before
```

原地执行

```python
z=torch.zeros_like(y)
print(F"z={z},id(z)={id(z)}")
z[:]=y+x
print(F"z={z},id(z)={id(z)}")
```

```python
before=id(x)
x+=y
id(x)==before
```
